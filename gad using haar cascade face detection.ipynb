Importing necessary packages

```python
import cv2
import argparse
```

Loading the pre-built model for facial detection.

```python
face_cascade = cv2.CascadeClassifier('/Users/admin/Desktop/project/haarcascade_frontalface_alt.xml')
```

creating an argument parser so we can get the image argument from the command prompt

```python
parser=argparse.ArgumentParser()
parser.add_argument('--image')
args, unknown = parser.parse_known_args()
```

Creating 3 separate lists to store Model_Mean_Values, Age and Gender.

```python
MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)
age_list = ['(0, 2)', '(4, 6)', '(8, 12)', '(15, 20)', '(25, 32)', '(38, 43)', '(48, 53)', '(60, 100)']
gender_list = ['Male', 'Female']
```

Initializing and loading the protocol buffer and model for age and gender.

```python
ageProto="/Users/admin/Desktop/gad/age_deploy.prototxt"
ageModel="/Users/admin/Desktop/gad/age_net.caffemodel"
genderProto="/Users/admin/Desktop/gad/gender_deploy.prototxt"
genderModel="/Users/admin/Desktop/gad/gender_net.caffemodel"
```

```python
age_net=cv2.dnn.readNet(ageModel,ageProto)
gender_net=cv2.dnn.readNet(genderModel,genderProto)
```

capturing video stream

```python
video=cv2.VideoCapture(args.image if args.image else 0)
padding=20
```

## Detecting age and gender

```python
while cv2.waitKey(1)<0:
    hasFrame,image=video.read()
    if not hasFrame:
        cv2.waitKey()
        break
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
        face_img = image[y:y+h, h:h+w].copy()
        blob = cv2.dnn.blobFromImage(face_img, 1, (227, 227), MODEL_MEAN_VALUES, swapRB=False)
        gender_net.setInput(blob)
        gender_preds = gender_net.forward()
        gender = gender_list[gender_preds[0].argmax()]
        print("Gender : " + gender)
        age_net.setInput(blob)
        age_preds = age_net.forward()
        age = age_list[age_preds[0].argmax()]
        print("Age Range: " + age)
        overlay_text = "%s %s" % (gender, age)
        cv2.putText(image, overlay_text, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.imshow('img', image)
```
